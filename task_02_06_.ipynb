Даны вещественные числа a, b, c (a≠0).
Решите уравнение ax2+bx+c=0. При выводе значений оставьте 1 знак после запятой.

```python
# Задание task_02_06.

a = float(input())
b = float(input())
c = float(input())

D = b**2 - 4*a*c   #находим дискриминант

if D < 0:
  print('Решений нет')
elif D == 0:
  i = (-b)/(2*a)
  print(round(i,1))
elif D > 0:
  x_1 = (-b - D**(1/2))/(2*a)
  x_2 = (-b + D**(1/2))/(2*a)
  x_1, x_2 = round(x_1,1), round(x_2,1)
  print(f'''x1 = {x_1}, x2 = {x_2}''')
# --------------
# Пример вывода:
#
# a = 1
# b = 2
# c = 3
# Решений нет
#
# a = 1
# b = 2
# c = 1
# x = -1.0
#
# a = 1
# b = 3
# c = -4
# x1 = -4.0, x2 = 1.0
```
