Дан список из чисел.
Определите их НОК (наименьшее общее кратное) и НОД (наибольший общий
делитель).

```python
def gcd(first, second):    # определяем функцию для нахождения НОД двух чисел
    while second != 0:
        first, second = second, first % second
    return abs(first)


def lcm(first, second):    # определяем функцию для нахождения НОК двух чисел
    return abs(first * second) // gcd(first, second)


def gcd_nums(nums):    # функция для вычисления НОД для списка чисел
    result = nums[0]
    for num in nums[1:]:
        result = gcd(result, num)
    return result


def lcm_nums(nums):   # функция для вычисления НОК для списка чисел
    result = nums[0]
    for num in nums[1:]:
        result = lcm(result, num)
    return result


nums = list(map(int, input().split()))

print("НОД =", gcd_nums(nums))
print("НОК =", lcm_nums(nums))


# --------------
# Пример вывода:
#
# Введите числа через пробел: 8 10 14
# НОД = 2
# НОК = 280
#
# Введите числа через пробел: 6 8 24 16
# НОД = 2
# НОК = 48
```
